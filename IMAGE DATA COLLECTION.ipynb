```python
from bing_image_downloader import downloader




# Types of faceshape
face_shape=["Oval","Round","Square","Diamond","Heart","Pear","Oblong"]




# different faceshape in different folder
for i in range(len(face_shape)):
    downloader.download("Celebrities with "+face_shape[i]+" face shape in single image", limit=10,  output_dir='dataset '+str(face_shape[i]))
```

```python
from keras.preprocessing.image import ImageDataGenerator,array_to_img,img_to_array ,load_img
```

```python
datagen = ImageDataGenerator(rotation_range=10, # rotation
        width_shift_range=0.2, # horizontal shift
        height_shift_range=0.2, # vertical shift
        zoom_range=0.2, # zoom
        horizontal_flip=True, # horizontal flip
        brightness_range=[0.2,1.2] )

img = load_img(r"C:\Users\bharat natrayn\Desktop\YOU TUBE\DATA COLLECTION\datasetDiamond\Celebrities with Diamond face shape in single image\Image_1.jpg")

x=img_to_array(img)

x=x.reshape((1,) + x.shape)

x.shape
```

```python
i=0
for batch in datagen.flow(x,batch_size=1,save_to_dir=r"C:\Users\bharat natrayn\Desktop\YOU TUBE\DATA COLLECTION\result"):
    i+=1
    if i>20:
        break
```
